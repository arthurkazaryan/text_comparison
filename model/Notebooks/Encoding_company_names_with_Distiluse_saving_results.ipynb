```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/distiluse-base-multilingual-cased-v2')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
res_1 = model.encode(df['name_1'].tolist(), device='cuda')
```

```python
res_1_new = []
for row in res_1:
  res_1_new.append([row])
```

```python
df['name_1_vec'] = res_1_new
```

```python
%%time
res_2 = model.encode(df['name_2'].tolist(), device='cuda')
```

```python
def save_feather(df, path="dummy.feather"):
    df = df.reset_index()
    df.to_feather(path)


def load_feather(path="dummy.feather"):
    df = pd.read_feather(path)
    df.set_index("index", inplace=True)
    df.index.name = None
    return df
```

```python
%%time
save_feather(df, '/content/drive/MyDrive/123/df_vec_distiluse.feather')
```

```python
%%time
df_vec_distiluse = load_feather('/content/drive/MyDrive/123/df_vec_distiluse.feather')
```
